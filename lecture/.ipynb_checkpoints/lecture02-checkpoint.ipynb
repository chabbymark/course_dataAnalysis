## 2.1 数据类型

```python
type(10)
```

### 2.1.1 基本数据类型

#### * **数值类型 -- 整数/浮点数/复数**

**1. 整型(int)** 
- 带符号(正负, singed)
- 不限长度

```python
1 - 9
```

```python
2 ** 10
```

```python
2 ** 100000
```

- 双精度(64-bit)
- 15分位精度
- 最大值 ~$1.80\times 10^{308}$

两种浮点型数据格式: 普通计数法与科学计数法

```python
3.14159
```

```python
6.02e23
```

浮点型精度的影响

```python
print(1+1e-15)
print(1+1e-16)
```

```python
print(1.79e308)
print(1.81e308)
```

**3 复数(complex)**

两种复数定义格式

```python
100+3j
```

```python
complex(100, 3)
```

### 2.1.2 复合数据类型

#### * **序列类型 -- 列表/元组/域**

**1 列表(list)**

Python的列表(list)形式上类似于C语言的数组，但内部元素可以为任意类型。<br>
列表形式上用`[]`定义，并用`,`分隔不同元素:

```python
list = [123, 'abc', 3.14, True, None]
```

为什么需要列表?

- 需要用不同的表达方法定义同一类元素的集合，比如同时利用文本和RGB定义颜色:

```python
color = ['red', 'blue', [255, 0, 255]]
```

- 需用存储单个对象的不同属性，比如个人信息:

```python
person = ['John', '2001/07/14', '370305XXXXXX']
```

使用切片(slice)运算符`[]`访问列表元素，列表的**索引指数由0开始**

```python
print(list[0])
print(list[3])
print(list[5])
```

*思考：为什么会报错？*

列表可以同时访问多个元素

其中切片运算符`[n1:n2]`可以连续访问一段列表数据，所取得数据不包含元素`n2`

```python
print(list[1:4])
print(list[1], list[2], list[3], list[4])
```

切片运算符`[n1:n2:d]`可以以固定间隔`d`访问列表, 且`d`可以为负数(反向访问列表)

```python
print(list[0:4:2])
print(list[4:0:-2])
```

`n1`为0或者`n2`为列表最后一维时可以省略

```python
print(list[:3] == list[0:3])
print(list[2:5] == list[2:])
```

列表索引指数`n`可以为负数，当为负数时，表示从列表末尾开始第|`n`|个元素

```python
print(list[-1]==list[4])
print(list[-2]==list[3])
print(list[:4]==list[0:-1])
```

列表元素值可以改变

```python
print(list[1])
list[1] = 'def'
print(list[1])
```

**2 元组(tuple)**

元组(tuple)形式上与操作上与列表类似，用`()`定义

```python
tuple = (123, 'abc', 3.14, True, None)
```

**但元组元素不可更改**

```python
tuple[2] = 'def'
```

元组适用于**逻辑上**不可变更数值的数据类型，比如

```python
red = (255, 0, 0)
blue = (0, 0, 255)
pink = (255, 153, 153)
```

**3 域(range)**

域(range)数据主要用于定义整数型等差数列，但并不对每个元素进行储存，一般用于for循环的循环范围。

域的定义方法有三种: `range(n)`, `range(n1, n2)`, `range(n1, n2, d)`

```python
r1 = range(10) # 输入参数为(n)位时，元素默认从0开始, n-1结束，且公差为1
r2 = range(1, 9) # 输入参数为(n1, n2)时，数列由n1开始，n2-1结束，且公差为1
r3 = range(2, 8, 2) # 输入参数为(n1, n2, d)时，数列由n1开始，n2-1结束，且公差为d
print(r1)
print(r2)
print(r3)
```

```python
sum = 0
for ele in range(1, 101):
    sum += ele
print(sum)
```

#### * **集合类型(set)**

采用`{}`定义，重复元素只计算一次，类似于数学集合的概念。

```python
set = {123, 'abc', 3.14, True, None, 123, 'abc'}
print(set)
```

#### * **映射类型 -- 字典(dict)**

字典(dict)与集合一样均采用`{}`定义，但需要指明映射关系(键 &rarr; 值, key &rarr; value)，且字典是与集合完全不同的数据类型。

```python
person_list = ['John', '2001/07/14', '370305XXXXXX']
person_dict = {'name': 'John', 'birth': '2001/07/14', 'ID': '370305XXXXXX'}
```

字典单个元素的(key &rarr; value)映射类型由`:`定义，例如`'name': 'John'`。

**字典只能采用键(key)进行访问**

```python
print(person_dict['birth'])
```

```python
print(person_dict[1])
```
