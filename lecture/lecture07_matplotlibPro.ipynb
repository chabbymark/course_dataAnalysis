## 💡 运用`GridSpec()`函数实现更高维度的子图的网格化分配。

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=figsize)

### 关键步骤: add_gridspec(row, column) or GridSpec(row, column) ###
gs = fig.add_gridspec(2, 2) # 或者 gs = GridSpec(2, 2)
### 关键步骤: add_gridspec(row, column) or GridSpec(row, column) ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

### ❓ 似乎还变麻烦了

### 💡 现在才是开始

```python
## 比如我们可以利用add_gridspec改变子图所占横纵比例

fig = plt.figure(figsize=figsize)

### 关键步骤: width_ratios=(r1, r2), height_ratio=(r3, r4) ###
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1))
### 关键步骤: width_ratios=(r1, r2), height_ratio=(r3, r4) ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

```python
## 比如我们可以利用add_gridspec改变子图占整个画布的比例

fig = plt.figure(figsize=figsize)

### 关键步骤: left, right, bottom, top四个输入变量 ###
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.1, top=0.98)
### 关键步骤: left, right, bottom, top四个输入变量 ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

```python
## 比如我们可以利用add_gridspec改变子图的间距

fig = plt.figure(figsize=figsize)

### 关键步骤: hspace(height space) 与 wspace (width space) 两个输入变量 ###
gs = fig.add_gridspec(2, 2, width_ratios=(3, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.02, top=0.98,
    hspace=0.1, wspace=0.1)
### 关键步骤: hspace(height space) 与 wspace (width space) 两个输入变量 ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

```python
## 比如以上关于子图边界功能均可以用gridspce对象(本示例为gs)的update函数进行更新

fig = plt.figure(figsize=figsize)

### 关键步骤: gs.update()函数 ###
gs = fig.add_gridspec(2, 2, width_ratios=(3, 1), height_ratios=(2, 1))
gs.update(left=0.02, right=0.98, bottom=0.02, top=0.98,
    hspace=0.1, wspace=0.05)
### 关键步骤: gs.update()函数 ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

## 🗂 画图与格式进阶

```python
# 数据准备
import numpy as np
data = np.load("./attachment/ocean_windWave.npz")
x3d = data['x'] + 1.5 # 调整一下坐标的水平中心
y3d = data['y'] + 1.5 # 调整一下坐标的水平中心
z3d = data['z']
w3d = data['w']

# 三个界面的位置，如果海洋上边界是一个长方体，三个截面类似于长方体三个互相垂直的面。
ix_crs = -1
iy_crs = 0
iz_crs = 5
```

```python
## 载入模块
from matplotlib import cm # colormap子模块



## 画图与构图
fig = plt.figure(figsize=figsize)
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.1, top=0.98, hspace=0.1, wspace=0.1)



## 画图准备
cmap_w = cm.get_cmap('RdYlBu_r') # 选择颜色图
levels_w = np.linspace(-0.02, 0.02, 21) # 设置等值线



## 画w数据的海表水平截面图
ax = fig.add_subplot(gs[0, 0])
cax = ax.contourf(x3d[iz_crs, :, :], y3d[iz_crs, :, :], w3d[iz_crs, :, :],
    levels=levels_w, cmap=cmap_w)

# 格式调整
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')



## 画w数据的xz截面图
ax = fig.add_subplot(gs[1, 0])
ax.contourf(x3d[:, iy_crs, :], z3d[:, iy_crs, :], w3d[:, iy_crs, :],
    levels=levels_w, cmap=cmap_w)

# 格式调整
ax.set_xlabel('x (m)')
ax.set_ylabel('z (m)')



## 画w数据yz截面
ax = fig.add_subplot(gs[0, 1])
ax.contourf(z3d[:, :, ix_crs], y3d[:, :, ix_crs], w3d[:, :, ix_crs],
    levels=levels_w, cmap=cmap_w)

# 格式调整
ax.set_xlabel('z (m)')
ax.set_ylabel('x (m)')
ax.invert_xaxis() # 反转x坐标



## 添加色阶图
cbar_ax = plt.axes([0.92, 0.3, 0.05, 0.4]) 
fig.colorbar(cax, cax=cbar_ax, ticks=levels_w)
```

### ❓ 等值线图超出量程
### &rarr; 💡 `ax.contourf()`的输入变量`extend`

### ❓ 色阶图位置错位且刻度过多
### &rarr; 💡 `plt.axes()`与`fig.colorbar()`

### ❓ 坐标轴互相覆盖且刻度混乱
### &rarr; 💡 `ax.set_xticklabels()`、`ax.set_yticklabels()`、`ax.xaxis.set_major_locator()`、`ax.yaxis.set_major_locator()`

### ❓ 坐标轴刻度比例可能有问题
### &rarr; 💡 `ax.set_aspect()`

## 💡 最终状态

```python
## 载入模块
from matplotlib.ticker import MultipleLocator # 刻度间隔模块
from matplotlib import cm # colormap子模块



## 刻度设置
xspacing = 30
yspacing = 30
zspacing = 15
xmaxLocator = MultipleLocator(xspacing)
ymaxLocator = MultipleLocator(yspacing)
zmaxLocator = MultipleLocator(zspacing)



## 坐标轴区域设置
xlims = (0, 150)
ylims = (0, 150)
zlims = (-90, 0)



## 画图与构图
fig = plt.figure(figsize=figsize)
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.1, top=0.98, hspace=0.1, wspace=0.1)



## 画图准备
cmap_w = cm.get_cmap('RdYlBu_r') # 选择颜色图: https://matplotlib.org/stable/tutorials/colors/colormaps.html
levels_w = np.linspace(-0.02, 0.02, 21) # 设置等值线



## 画w数据的海表水平截面图
ax = fig.add_subplot(gs[0, 0])
ax.set_aspect('equal') # 设置子图横纵坐标比例尺

cax = ax.contourf(x3d[iz_crs, :, :], y3d[iz_crs, :, :], w3d[iz_crs, :, :],
    levels=levels_w, cmap=cmap_w, extend='both') # extend可以扩展色阶量程

# 格式调整
ax.set_ylabel('y (m)')
ax.set_xticklabels([]) # 将刻度值隐藏
ax.xaxis.set_major_locator(xmaxLocator) # 设置刻度间隔
ax.yaxis.set_major_locator(ymaxLocator) # 设置刻度间隔
ax.set_xlim(xlims) # 设置坐标轴范围
ax.set_ylim(ylims) # 设置坐标轴范围



## 画w数据的xz截面图
ax = fig.add_subplot(gs[1, 0])
ax.set_aspect(0.84)

ax.contourf(x3d[:, iy_crs, :], z3d[:, iy_crs, :], w3d[:, iy_crs, :],
    levels=levels_w, cmap=cmap_w, extend='both')

# 格式调整
ax.set_xlabel('x (m)')
ax.set_ylabel('z (m)')
ax.xaxis.set_major_locator(xmaxLocator)
ax.yaxis.set_major_locator(zmaxLocator)
ax.set_xlim(xlims)
ax.set_ylim(zlims)



## 画w数据yz截面
ax = fig.add_subplot(gs[0, 1])
ax.contourf(z3d[:, :, ix_crs], y3d[:, :, ix_crs], w3d[:, :, ix_crs],
    levels=levels_w, cmap=cmap_w, extend='both')

# 格式调整
ax.set_xlabel('z (m)')
ax.set_yticklabels([])
ax.xaxis.set_major_locator(zmaxLocator)
ax.yaxis.set_major_locator(ymaxLocator)
ax.set_xlim(zlims)
ax.set_ylim(ylims)
ax.invert_xaxis() # 反转x坐标



## 添加色阶图
cbar_ax = plt.axes([0.82, 0.1, 0.05, 0.2]) # 建立一个新的子区域，格式是[x起始, y起始, 宽度, 高度]，单位是整个画布的归一化坐标。
cbar = fig.colorbar(cax, cax=cbar_ax, ticks=levels_w[::5])
cbar.ax.set_title("w (m/s)", ha='center', va='bottom')



## 保存图片
fig.savefig("./image/w_crs.png", dpi=300)
```
