## üí° ËøêÁî®`GridSpec()`ÂáΩÊï∞ÂÆûÁé∞Êõ¥È´òÁª¥Â∫¶ÁöÑÂ≠êÂõæÁöÑÁΩëÊ†ºÂåñÂàÜÈÖç„ÄÇ

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=figsize)

### ÂÖ≥ÈîÆÊ≠•È™§: add_gridspec(row, column) or GridSpec(row, column) ###
gs = fig.add_gridspec(2, 2) # ÊàñËÄÖ gs = GridSpec(2, 2)
### ÂÖ≥ÈîÆÊ≠•È™§: add_gridspec(row, column) or GridSpec(row, column) ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

### ‚ùì ‰ºº‰πéËøòÂèòÈ∫ªÁÉ¶‰∫Ü

### üí° Áé∞Âú®ÊâçÊòØÂºÄÂßã

```python
## ÊØîÂ¶ÇÊàë‰ª¨ÂèØ‰ª•Âà©Áî®add_gridspecÊîπÂèòÂ≠êÂõæÊâÄÂç†Ê®™Á∫µÊØî‰æã

fig = plt.figure(figsize=figsize)

### ÂÖ≥ÈîÆÊ≠•È™§: width_ratios=(r1, r2), height_ratio=(r3, r4) ###
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1))
### ÂÖ≥ÈîÆÊ≠•È™§: width_ratios=(r1, r2), height_ratio=(r3, r4) ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

```python
## ÊØîÂ¶ÇÊàë‰ª¨ÂèØ‰ª•Âà©Áî®add_gridspecÊîπÂèòÂ≠êÂõæÂç†Êï¥‰∏™ÁîªÂ∏ÉÁöÑÊØî‰æã

fig = plt.figure(figsize=figsize)

### ÂÖ≥ÈîÆÊ≠•È™§: left, right, bottom, topÂõõ‰∏™ËæìÂÖ•ÂèòÈáè ###
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.1, top=0.98)
### ÂÖ≥ÈîÆÊ≠•È™§: left, right, bottom, topÂõõ‰∏™ËæìÂÖ•ÂèòÈáè ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

```python
## ÊØîÂ¶ÇÊàë‰ª¨ÂèØ‰ª•Âà©Áî®add_gridspecÊîπÂèòÂ≠êÂõæÁöÑÈó¥Ë∑ù

fig = plt.figure(figsize=figsize)

### ÂÖ≥ÈîÆÊ≠•È™§: hspace(height space) ‰∏é wspace (width space) ‰∏§‰∏™ËæìÂÖ•ÂèòÈáè ###
gs = fig.add_gridspec(2, 2, width_ratios=(3, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.02, top=0.98,
    hspace=0.1, wspace=0.1)
### ÂÖ≥ÈîÆÊ≠•È™§: hspace(height space) ‰∏é wspace (width space) ‰∏§‰∏™ËæìÂÖ•ÂèòÈáè ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

```python
## ÊØîÂ¶Ç‰ª•‰∏äÂÖ≥‰∫éÂ≠êÂõæËæπÁïåÂäüËÉΩÂùáÂèØ‰ª•Áî®gridspceÂØπË±°(Êú¨Á§∫‰æã‰∏∫gs)ÁöÑupdateÂáΩÊï∞ËøõË°åÊõ¥Êñ∞

fig = plt.figure(figsize=figsize)

### ÂÖ≥ÈîÆÊ≠•È™§: gs.update()ÂáΩÊï∞ ###
gs = fig.add_gridspec(2, 2, width_ratios=(3, 1), height_ratios=(2, 1))
gs.update(left=0.02, right=0.98, bottom=0.02, top=0.98,
    hspace=0.1, wspace=0.05)
### ÂÖ≥ÈîÆÊ≠•È™§: gs.update()ÂáΩÊï∞ ###

ax_ul = fig.add_subplot(gs[0, 0]) # ul = upper left
ax_ur = fig.add_subplot(gs[0, 1]) # ur = upper right
ax_ll = fig.add_subplot(gs[1, 0]) # ll = lower left
ax_lr = fig.add_subplot(gs[1, 1]) # lr = lower right
```

## üóÇ ÁîªÂõæ‰∏éÊ†ºÂºèËøõÈò∂

```python
# Êï∞ÊçÆÂáÜÂ§á
import numpy as np
data = np.load("./attachment/ocean_windWave.npz")
x3d = data['x'] + 1.5 # Ë∞ÉÊï¥‰∏Ä‰∏ãÂùêÊ†áÁöÑÊ∞¥Âπ≥‰∏≠ÂøÉ
y3d = data['y'] + 1.5 # Ë∞ÉÊï¥‰∏Ä‰∏ãÂùêÊ†áÁöÑÊ∞¥Âπ≥‰∏≠ÂøÉ
z3d = data['z']
w3d = data['w']

# ‰∏â‰∏™ÁïåÈù¢ÁöÑ‰ΩçÁΩÆÔºåÂ¶ÇÊûúÊµ∑Ê¥ã‰∏äËæπÁïåÊòØ‰∏Ä‰∏™ÈïøÊñπ‰ΩìÔºå‰∏â‰∏™Êà™Èù¢Á±ª‰ºº‰∫éÈïøÊñπ‰Ωì‰∏â‰∏™‰∫íÁõ∏ÂûÇÁõ¥ÁöÑÈù¢„ÄÇ
ix_crs = -1
iy_crs = 0
iz_crs = 5
```

```python
## ËΩΩÂÖ•Ê®°Âùó
from matplotlib import cm # colormapÂ≠êÊ®°Âùó



## ÁîªÂõæ‰∏éÊûÑÂõæ
fig = plt.figure(figsize=figsize)
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.1, top=0.98, hspace=0.1, wspace=0.1)



## ÁîªÂõæÂáÜÂ§á
cmap_w = cm.get_cmap('RdYlBu_r') # ÈÄâÊã©È¢úËâ≤Âõæ
levels_w = np.linspace(-0.02, 0.02, 21) # ËÆæÁΩÆÁ≠âÂÄºÁ∫ø



## ÁîªwÊï∞ÊçÆÁöÑÊµ∑Ë°®Ê∞¥Âπ≥Êà™Èù¢Âõæ
ax = fig.add_subplot(gs[0, 0])
cax = ax.contourf(x3d[iz_crs, :, :], y3d[iz_crs, :, :], w3d[iz_crs, :, :],
    levels=levels_w, cmap=cmap_w)

# Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')



## ÁîªwÊï∞ÊçÆÁöÑxzÊà™Èù¢Âõæ
ax = fig.add_subplot(gs[1, 0])
ax.contourf(x3d[:, iy_crs, :], z3d[:, iy_crs, :], w3d[:, iy_crs, :],
    levels=levels_w, cmap=cmap_w)

# Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('x (m)')
ax.set_ylabel('z (m)')



## ÁîªwÊï∞ÊçÆyzÊà™Èù¢
ax = fig.add_subplot(gs[0, 1])
ax.contourf(z3d[:, :, ix_crs], y3d[:, :, ix_crs], w3d[:, :, ix_crs],
    levels=levels_w, cmap=cmap_w)

# Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('z (m)')
ax.set_ylabel('x (m)')
ax.invert_xaxis() # ÂèçËΩ¨xÂùêÊ†á



## Ê∑ªÂä†Ëâ≤Èò∂Âõæ
cbar_ax = plt.axes([0.92, 0.3, 0.05, 0.4]) 
fig.colorbar(cax, cax=cbar_ax, ticks=levels_w)
```

### ‚ùì Á≠âÂÄºÁ∫øÂõæË∂ÖÂá∫ÈáèÁ®ã
### &rarr; üí° `ax.contourf()`ÁöÑËæìÂÖ•ÂèòÈáè`extend`

### ‚ùì Ëâ≤Èò∂Âõæ‰ΩçÁΩÆÈîô‰Ωç‰∏îÂàªÂ∫¶ËøáÂ§ö
### &rarr; üí° `plt.axes()`‰∏é`fig.colorbar()`

### ‚ùì ÂùêÊ†áËΩ¥‰∫íÁõ∏Ë¶ÜÁõñ‰∏îÂàªÂ∫¶Ê∑∑‰π±
### &rarr; üí° `ax.set_xticklabels()`„ÄÅ`ax.set_yticklabels()`„ÄÅ`ax.xaxis.set_major_locator()`„ÄÅ`ax.yaxis.set_major_locator()`

### ‚ùì ÂùêÊ†áËΩ¥ÂàªÂ∫¶ÊØî‰æãÂèØËÉΩÊúâÈóÆÈ¢ò
### &rarr; üí° `ax.set_aspect()`

## üí° ÊúÄÁªàÁä∂ÊÄÅ

```python
## ËΩΩÂÖ•Ê®°Âùó
from matplotlib.ticker import MultipleLocator # ÂàªÂ∫¶Èó¥ÈöîÊ®°Âùó
from matplotlib import cm # colormapÂ≠êÊ®°Âùó



## ÂàªÂ∫¶ËÆæÁΩÆ
xspacing = 30
yspacing = 30
zspacing = 15
xmaxLocator = MultipleLocator(xspacing)
ymaxLocator = MultipleLocator(yspacing)
zmaxLocator = MultipleLocator(zspacing)



## ÂùêÊ†áËΩ¥Âå∫ÂüüËÆæÁΩÆ
xlims = (0, 150)
ylims = (0, 150)
zlims = (-90, 0)



## ÁîªÂõæ‰∏éÊûÑÂõæ
fig = plt.figure(figsize=figsize)
gs = fig.add_gridspec(2, 2, width_ratios=(2, 1), height_ratios=(2, 1),
    left=0.1, right=0.98, bottom=0.1, top=0.98, hspace=0.1, wspace=0.1)



## ÁîªÂõæÂáÜÂ§á
cmap_w = cm.get_cmap('RdYlBu_r') # ÈÄâÊã©È¢úËâ≤Âõæ: https://matplotlib.org/stable/tutorials/colors/colormaps.html
levels_w = np.linspace(-0.02, 0.02, 21) # ËÆæÁΩÆÁ≠âÂÄºÁ∫ø



## ÁîªwÊï∞ÊçÆÁöÑÊµ∑Ë°®Ê∞¥Âπ≥Êà™Èù¢Âõæ
ax = fig.add_subplot(gs[0, 0])
ax.set_aspect('equal') # ËÆæÁΩÆÂ≠êÂõæÊ®™Á∫µÂùêÊ†áÊØî‰æãÂ∞∫

cax = ax.contourf(x3d[iz_crs, :, :], y3d[iz_crs, :, :], w3d[iz_crs, :, :],
    levels=levels_w, cmap=cmap_w, extend='both') # extendÂèØ‰ª•Êâ©Â±ïËâ≤Èò∂ÈáèÁ®ã

# Ê†ºÂºèË∞ÉÊï¥
ax.set_ylabel('y (m)')
ax.set_xticklabels([]) # Â∞ÜÂàªÂ∫¶ÂÄºÈöêËóè
ax.xaxis.set_major_locator(xmaxLocator) # ËÆæÁΩÆÂàªÂ∫¶Èó¥Èöî
ax.yaxis.set_major_locator(ymaxLocator) # ËÆæÁΩÆÂàªÂ∫¶Èó¥Èöî
ax.set_xlim(xlims) # ËÆæÁΩÆÂùêÊ†áËΩ¥ËåÉÂõ¥
ax.set_ylim(ylims) # ËÆæÁΩÆÂùêÊ†áËΩ¥ËåÉÂõ¥



## ÁîªwÊï∞ÊçÆÁöÑxzÊà™Èù¢Âõæ
ax = fig.add_subplot(gs[1, 0])
ax.set_aspect(0.84)

ax.contourf(x3d[:, iy_crs, :], z3d[:, iy_crs, :], w3d[:, iy_crs, :],
    levels=levels_w, cmap=cmap_w, extend='both')

# Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('x (m)')
ax.set_ylabel('z (m)')
ax.xaxis.set_major_locator(xmaxLocator)
ax.yaxis.set_major_locator(zmaxLocator)
ax.set_xlim(xlims)
ax.set_ylim(zlims)



## ÁîªwÊï∞ÊçÆyzÊà™Èù¢
ax = fig.add_subplot(gs[0, 1])
ax.contourf(z3d[:, :, ix_crs], y3d[:, :, ix_crs], w3d[:, :, ix_crs],
    levels=levels_w, cmap=cmap_w, extend='both')

# Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('z (m)')
ax.set_yticklabels([])
ax.xaxis.set_major_locator(zmaxLocator)
ax.yaxis.set_major_locator(ymaxLocator)
ax.set_xlim(zlims)
ax.set_ylim(ylims)
ax.invert_xaxis() # ÂèçËΩ¨xÂùêÊ†á



## Ê∑ªÂä†Ëâ≤Èò∂Âõæ
cbar_ax = plt.axes([0.82, 0.1, 0.05, 0.2]) # Âª∫Á´ã‰∏Ä‰∏™Êñ∞ÁöÑÂ≠êÂå∫ÂüüÔºåÊ†ºÂºèÊòØ[xËµ∑Âßã, yËµ∑Âßã, ÂÆΩÂ∫¶, È´òÂ∫¶]ÔºåÂçï‰ΩçÊòØÊï¥‰∏™ÁîªÂ∏ÉÁöÑÂΩí‰∏ÄÂåñÂùêÊ†á„ÄÇ
cbar = fig.colorbar(cax, cax=cbar_ax, ticks=levels_w[::5])
cbar.ax.set_title("w (m/s)", ha='center', va='bottom')



## ‰øùÂ≠òÂõæÁâá
fig.savefig("./image/w_crs.png", dpi=300)
```
