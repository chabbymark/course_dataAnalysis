```python
# 0. ËôöÊãüÊï∞ÊçÆ
import numpy as np
x_data = np.arange(-5, 5.01, 0.25)
y_data = x_data ** 2

# 1. ÂàõÂª∫ÁîªÂ∏É‰∏éÊûÑÂõæ
fig = plt.figure() # ‰∏éÊâìÂºÄÊñá‰ª∂ËøîÂõûÊñá‰ª∂ÂØπË±°Á±ª‰ººÔºåËøôÈáå‰ºöËøîÂõû‰∏Ä‰∏™ÂõæÂÉèÂØπË±°
ax = fig.subplots() # Ëé∑ÂæóÁîªÂõæÂå∫ÂüüÁöÑÂùêÊ†áÔºåÈªòËÆ§ÊòØÂè™Êúâ1Ë°å1Âàó

# 2. ÁîªÂõæ
ax.plot(x_data, y_data, color='C0', linestyle='--') # plotÊòØÁ∫øÂõæÔºåËøòÊúâÂ§öÁßçÁ±ªÂûãÂèØÈÄâ

# 3. Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel("x") # ‰∏Ä‰∫õÂùêÊ†áËΩ¥Ê†áËØÜ
ax.set_ylabel("y")

# 4. ÂõæÁâáÂ±ïÁ§∫‰∏é‰øùÂ≠ò
plt.show() # ÂçïÁ∫ØÁöÑÂ±ïÁ§∫ÁªìÊûú
```

## üí° Êï£ÁÇπÂõæ(scatter)

```python
# 0. Êï∞ÊçÆËé∑Âèñ
data_ocean = np.loadtxt("./attachment/ocean_density.txt", skiprows=1)
depth = data_ocean[:, 0]
salinity = data_ocean[:, 1]
temperature = data_ocean[:, 2]
density = data_ocean[:, 3]

# 1. ÂàõÂª∫ÁîªÂ∏É‰∏éÊûÑÂõæ
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# 2. ÁîªÂõæ / 3. Ê†ºÂºèË∞ÉÊï¥
### ÁîªÂõæÈÉ®ÂàÜÂºÄÂßã ###
# Â≠êÂå∫Âüü 1
ax = axes[0]
ax.scatter(salinity, density, s=20, marker='o') # sÊåáÂÆöÁöÑÊòØÂ§ßÂ∞è(size)
ax.set_xlabel("Salinity (PSU)")
ax.set_ylabel("Density $\mathrm{(kg/m^3)}$") # ÊîØÊåÅLaTeXÊï∞Â≠¶ÂÖ¨Âºè(e.g., $\mathrm{(kg/m^3)}$)

ax = axes[1]
ax.scatter(temperature, density, s=20, marker='o', edgecolor='k', facecolor='y')
ax.set_xlabel("Temperature $\mathrm{(^\circ C)}$")
### ÁîªÂõæÈÉ®ÂàÜÁªìÊùü ###

# 4. ÂõæÁâáÂ±ïÁ§∫‰∏é‰øùÂ≠ò
plt.show()
```

## üí° Á≠âÂÄºÁ∫øÂõæ(contourf)

```python
## ËΩΩÂÖ•Êñ∞ÁöÑÊ®°Âùó
from matplotlib import cm # colormapÂ≠êÊ®°Âùó

# 0. Êï∞ÊçÆËé∑Âèñ - ÈÉ®ÂàÜ‰∫öÈ©¨ÈÄäÂú∞ÂΩ¢
data = np.load("./attachment/amazon_topo.npz")
dy, dx = (2, 2)
tpg = data['topo'] # Âú∞ÂΩ¢Êï∞ÊçÆ
x_gd = np.arange(0, dx*tpg.shape[1], dx) # xÂùêÊ†á
y_gd = np.arange(0, dy*tpg.shape[0], dy) # yÂùêÊ†á

# 1. ÂàõÂª∫ÁîªÂ∏É‰∏éÊûÑÂõæ
fig, ax = plt.subplots(figsize=(6, 6))

# 2. ÁîªÂõæ
### ÁîªÂõæÈÉ®ÂàÜÂºÄÂßã ###
cmap = cm.get_cmap('rainbow') # ÈÄâÊã©È¢úËâ≤Âõæ: https://matplotlib.org/stable/tutorials/colors/colormaps.html
levels = np.arange(0, 65, 5) # ËÆæÁΩÆÁ≠âÂÄºÁ∫ø

cax = ax.contourf(x_gd, y_gd, tpg, levels=levels, cmap=cmap)
### ÁîªÂõæÈÉ®ÂàÜÁªìÊùü ###

# 3. Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')

# 4. ÂõæÁâáÂ±ïÁ§∫‰∏é‰øùÂ≠ò
plt.show()
```

## üí° Áõ¥ÊñπÂõæ(hist)

```python
# 0. Êï∞ÊçÆËé∑Âèñ - ÈÉ®ÂàÜ‰∫öÈ©¨ÈÄäÂú∞ÂΩ¢
data = np.load("./attachment/amazon_topo.npz")
tpg = data['topo'] # Âú∞ÂΩ¢Êï∞ÊçÆ

# 1. ÂàõÂª∫ÁîªÂ∏É‰∏éÊûÑÂõæ
fig, ax = plt.subplots(figsize=(8, 4))

# 2. ÁîªÂõæ
### ÁîªÂõæÈÉ®ÂàÜÂºÄÂßã ###
bins = np.linspace(0, 60, 13)
cax = ax.hist(tpg.ravel(), bins=bins, density=True)
### ÁîªÂõæÈÉ®ÂàÜÁªìÊùü ###

# 3. Ê†ºÂºèË∞ÉÊï¥
ax.set_xlabel('elevation (m)')
ax.set_ylabel('PDF $(\mathrm{m^{-1}})$')

# 4. ÂõæÁâáÂ±ïÁ§∫‰∏é‰øùÂ≠ò
plt.show()
```

## üí° ËøòÊúâÊõ¥Â§öÁßçÁ±ªÁöÑÁîªÂõæÂáΩÊï∞ÔºåÂêåÂ≠¶‰ª¨‰∏ÄÂÆöË¶ÅÊ†πÊçÆËá™Â∑±ÁöÑÈúÄÊ±ÇÂéªÊü•ÈòÖ„ÄÇÁ§∫‰æãÁΩëÂùÄ:https://matplotlib.org/stable/gallery/index

## üóÇ Ê†ºÂºèË∞ÉÊï¥

## üí° ÂõûÂà∞Á≠âÂÄºÁ∫øÂõæ

```python
# ÂÖàÂÆö‰∏Ä‰∏™ÂáΩÊï∞Áî®Êù•Âø´ÈÄüÁîªÂõæ
def plot_tpg():
    # 0. Êï∞ÊçÆËé∑Âèñ - ÈÉ®ÂàÜ‰∫öÈ©¨ÈÄäÂú∞ÂΩ¢
    data = np.load("./attachment/amazon_topo.npz")
    dy, dx = (2, 2)
    tpg = data['topo'] # Âú∞ÂΩ¢Êï∞ÊçÆ
    x_gd = np.arange(0, dx*tpg.shape[1], dx) # xÂùêÊ†á
    y_gd = np.arange(0, dy*tpg.shape[0], dy) # yÂùêÊ†á

    # 1. ÂàõÂª∫ÁîªÂ∏É‰∏éÊûÑÂõæ
    fig, ax = plt.subplots(figsize=(6, 6))

    # 2. ÁîªÂõæ
    ### ÁîªÂõæÈÉ®ÂàÜÂºÄÂßã ###
    cmap = cm.get_cmap('rainbow') # ÈÄâÊã©È¢úËâ≤Âõæ: https://matplotlib.org/stable/tutorials/colors/colormaps.html
    levels = np.arange(0, 65, 5) # ËÆæÁΩÆÁ≠âÂÄºÁ∫ø

    cax = ax.contourf(x_gd, y_gd, tpg, levels=levels, cmap=cmap)
    ### ÁîªÂõæÈÉ®ÂàÜÁªìÊùü ###
    return fig, ax, cax
```

```python
# Âè™ÊòØÁîªÂõæ
fig, ax, cax = plot_tpg()
```

## ? ÂùêÊ†á‰ª£Ë°®‰ªÄ‰πàÂê´‰πâ
## üí° `set_xlabel()`‰∏é`set_ylabel()`ÂáΩÊï∞ËÆæÁΩÆÂùêÊ†áÊ†áËØÜ

```python
# ÂàùÂßãÁä∂ÊÄÅ
fig, ax, cax = plot_tpg()

# Ê∑ªÂä†ÂùêÊ†áÊ†áËØÜ
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
```

## ‚ùì ÂùêÊ†áËΩ¥ÂàªÂ∫¶Â§™Â§ßÊàñÂ§™Â∞èÊÄé‰πàÂäû
## üí° `set_xticks()`‰∏é`set_yticks()`ËÆæÁΩÆÂùêÊ†áÂàªÂ∫¶

```python
# ÂàùÂßãÁä∂ÊÄÅ
fig, ax, cax = plot_tpg()
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")

# ËÆæÁΩÆÂùêÊ†áÂàªÂ∫¶
xticks = np.arange(0, 3001, 1000)
yticks = np.arange(0, 3001, 1000)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
```

## ‚ùì È¢úËâ≤‰ª£Ë°®‰ªÄ‰πà
## üí° `colorbar()`ÂáΩÊï∞Ê∑ªÂä†È¢úËâ≤‰ø°ÊÅØ

```python
# ÂàùÂßãÁä∂ÊÄÅ
fig, ax, cax = plot_tpg()
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
xticks = np.arange(0, 3001, 1000)
yticks = np.arange(0, 3001, 1000)
ax.set_xticks(xticks)
ax.set_yticks(yticks)

# ËÆæÁΩÆËâ≤Èò∂Ê†è
levels = np.arange(0, 61, 10)
cbar_ax = plt.axes([0.92, 0.3, 0.05, 0.4]) # Âª∫Á´ã‰∏Ä‰∏™Êñ∞ÁöÑÂ≠êÂå∫ÂüüÔºåÊ†ºÂºèÊòØ[xËµ∑Âßã, yËµ∑Âßã, ÂÆΩÂ∫¶, È´òÂ∫¶]ÔºåÂçï‰ΩçÊòØÊï¥‰∏™ÁîªÂ∏ÉÁöÑÂΩí‰∏ÄÂåñÂùêÊ†á„ÄÇ
plt.colorbar(cax, cax=cbar_ax, ticks=levels)
```
