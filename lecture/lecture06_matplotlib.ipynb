```python
# 0. 虚拟数据
import numpy as np
x_data = np.arange(-5, 5.01, 0.25)
y_data = x_data ** 2

# 1. 创建画布与构图
fig = plt.figure() # 与打开文件返回文件对象类似，这里会返回一个图像对象
ax = fig.subplots() # 获得画图区域的坐标，默认是只有1行1列

# 2. 画图
ax.plot(x_data, y_data, color='C0', linestyle='--') # plot是线图，还有多种类型可选

# 3. 格式调整
ax.set_xlabel("x") # 一些坐标轴标识
ax.set_ylabel("y")

# 4. 图片展示与保存
plt.show() # 单纯的展示结果
```

## 🗂 画图

## ❓如果想多画几根线怎么办(plot)

## 💡 方法1:

```python
x_data = np.arange(-5, 5.01, 0.25)
y1_data = x_data ** 2
y2_data = 0.5* x_data ** 2 + 0.6 * x_data

fig, ax = plt.subplots(figsize=(8, 4))

### 画图部分开始 ###
ax.plot(x_data, y1_data, color='C0', linestyle='--')
ax.plot(x_data, y2_data, color='C1', linestyle='-.')
### 画图部分结束 ###

ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

## 💡 方法2:

```python
x_data = np.arange(-5, 5.01, 0.25)
y1_data = x_data ** 2
y2_data = 0.5* x_data ** 2 + 0.6 * x_data

fig, ax = plt.subplots(figsize=(8, 4))

### 画图部分开始 ###
ax.plot(x_data, y1_data, 'b-o', x_data, y2_data, 'r-.x')
# 注意到Matploblib也可以使用MATLAB类似的画图格式指定方式(e.g., 'b-o'与'r-.x')
# 这里的格式是‘<颜色><线型><标识>’
### 画图部分结束 ###

# 3. 格式调整
ax.set_xlabel("x")
ax.set_ylabel("y")

# 4. 图片展示与保存
plt.show()
```

## 💡 散点图(scatter)

```python
# 0. 数据获取
data_ocean = np.loadtxt("./attachment/ocean_density.txt", skiprows=1)
depth = data_ocean[:, 0]
salinity = data_ocean[:, 1]
temperature = data_ocean[:, 2]
density = data_ocean[:, 3]

# 1. 创建画布与构图
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# 2. 画图 / 3. 格式调整
### 画图部分开始 ###
# 子区域 1
ax = axes[0]
ax.scatter(salinity, density, s=20, marker='o') # s指定的是大小(size)
ax.set_xlabel("Salinity (PSU)")
ax.set_ylabel("Density $\mathrm{(kg/m^3)}$") # 支持LaTeX数学公式(e.g., $\mathrm{(kg/m^3)}$)

ax = axes[1]
ax.scatter(temperature, density, s=20, marker='o', edgecolor='k', facecolor='y')
ax.set_xlabel("Temperature $\mathrm{(^\circ C)}$")
### 画图部分结束 ###

# 4. 图片展示与保存
plt.show()
```

## 💡 等值线图(contourf)

```python
## 载入新的模块
from matplotlib import cm # colormap子模块

# 0. 数据获取 - 部分亚马逊地形
data = np.load("./attachment/amazon_topo.npz")
dy, dx = (2, 2)
tpg = data['topo'] # 地形数据
x_gd = np.arange(0, dx*tpg.shape[1], dx) # x坐标
y_gd = np.arange(0, dy*tpg.shape[0], dy) # y坐标

# 1. 创建画布与构图
fig, ax = plt.subplots(figsize=(6, 6))

# 2. 画图
### 画图部分开始 ###
cmap = cm.get_cmap('rainbow') # 选择颜色图: https://matplotlib.org/stable/tutorials/colors/colormaps.html
levels = np.arange(0, 65, 5) # 设置等值线

cax = ax.contourf(x_gd, y_gd, tpg, levels=levels, cmap=cmap)
### 画图部分结束 ###

# 3. 格式调整
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')

# 4. 图片展示与保存
plt.show()
```

## 🗂 格式调整

## 💡 回到等值线图

```python
# 先定一个函数用来快速画图
def plot_tpg():
    # 0. 数据获取 - 部分亚马逊地形
    data = np.load("./attachment/amazon_topo.npz")
    dy, dx = (2, 2)
    tpg = data['topo'] # 地形数据
    x_gd = np.arange(0, dx*tpg.shape[1], dx) # x坐标
    y_gd = np.arange(0, dy*tpg.shape[0], dy) # y坐标

    # 1. 创建画布与构图
    fig, ax = plt.subplots(figsize=(6, 6))

    # 2. 画图
    ### 画图部分开始 ###
    cmap = cm.get_cmap('rainbow') # 选择颜色图: https://matplotlib.org/stable/tutorials/colors/colormaps.html
    levels = np.arange(0, 65, 5) # 设置等值线

    cax = ax.contourf(x_gd, y_gd, tpg, levels=levels, cmap=cmap)
    ### 画图部分结束 ###
    return fig, ax, cax
```

```python
# 只是画图
fig, ax, cax = plot_tpg()
```

## ? 坐标代表什么含义
## 💡 `set_xlabel()`与`set_ylabel()`函数设置坐标标识

```python
# 初始状态
fig, ax, cax = plot_tpg()

# 添加坐标标识
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
```

## ❓ 坐标轴刻度太大或太小怎么办
## 💡 `set_xticks()`与`set_yticks()`设置坐标刻度

```python
# 初始状态
fig, ax, cax = plot_tpg()
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")

# 设置坐标刻度
xticks = np.arange(0, 3001, 1000)
yticks = np.arange(0, 3001, 1000)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
```

## ❓ 颜色代表什么
## 💡 `colorbar()`函数添加颜色信息

```python
# 初始状态
fig, ax, cax = plot_tpg()
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
xticks = np.arange(0, 3001, 1000)
yticks = np.arange(0, 3001, 1000)
ax.set_xticks(xticks)
ax.set_yticks(yticks)

# 设置色阶栏
levels = np.arange(0, 61, 10)
cbar_ax = plt.axes([0.92, 0.3, 0.05, 0.4]) # 建立一个新的子区域，格式是[x起始, y起始, 宽度, 高度]，单位是整个画布的归一化坐标。
plt.colorbar(cax, cax=cbar_ax, ticks=levels)
```

## 💡 回到线图

```python
# 初始状态
x_data = np.arange(-5, 5.01, 0.25)
y1_data = x_data ** 2
y2_data = 0.5* x_data ** 2 + 0.6 * x_data

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x_data, y1_data, color='C0', linestyle='--')
ax.plot(x_data, y2_data, color='C1', linestyle='-.')

ax.set_xlabel("x")
ax.set_ylabel("y")
```

## ❓怎么给两条线加说明
## 💡画图时指定`label`标量，并用`legend()`函数生成说明

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x_data, y1_data, color='C0', linestyle='--', label=r"$y=x^2$")
ax.plot(x_data, y2_data, color='C1', linestyle='-.', label=r"$y=0.5x^2+0.6x$")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc="lower right")
```

## 💡 `savefig()`函数

```python
# 完整图片
fig, ax, cax = plot_tpg()
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
xticks = np.arange(0, 3001, 1000)
yticks = np.arange(0, 3001, 1000)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
levels = np.arange(0, 61, 10)
cbar_ax = plt.axes([0.92, 0.3, 0.05, 0.4]) # 建立一个新的子区域，格式是[x起始, y起始, 宽度, 高度]，单位是整个画布的归一化坐标。
plt.colorbar(cax, cax=cbar_ax, ticks=levels)

# 4. 图片显示与储存
fig.savefig("./attachment/topography.png", dpi=300)
# 第一个变量为文件名，dpi变量标识精度(Dots Per Inch, 每英寸的像素)
```
