- **路径、目录、文件？**
    - 路径(path): 系统中文件、目录及设备等的**地址**。
    - 目录(directory): 文件夹。
    - 文件(file): 数据流集合。

- **路径知识扩展**
1. 绝对路径与相对路径
- 绝对路径(absolute path): 由Linux根目录开始定义的路径，不会有歧义
    - 类比：在京东上买东西写的地址
        - /福建省/厦门市/翔安区/翔安南路/4221号/文宣楼/B312/陈先生
        - **这里的根目录“/”是“中国”**
        - 任何人都可以根据这个地址寄送物品

- 相对路径(relative path): 由Linux当前目录或者上级目录开始定义的路径，写法简洁
    - 类比1: 老师临时通知你们上机的地点
        - ./希平楼/B2/207 (这里的“.”指代翔安南路4221号或厦门大学翔安校区)
        - 注: *如过老师通知你们的地点是“/福建省/厦门市/翔安区/翔安南路/4221号/希平楼/B2/207”，你们会怎么想？*
    - 类比2: 今晚约朋友去沙美村黄记麻辣香锅吃饭
        - 绝对路径: /福建省/厦门市/翔安区/沙美南里/302-106黄记麻辣香锅
        - 当前路径: /福建省/厦门市/翔安区/翔安南路/4221号
        - 相对路径: ../../沙美南里/302-106黄记麻辣香锅
    - 注: **使用当前路径下的目录或文件时，"./"可省略**

2. 特殊路径简写
- 当前目录: "."
- 上一级目录: ".."
- 家目录: "~"
- 进入的上一个目录: "-"
- 练习: 试试`cd .`、`cd ..`、`cd ~`、`cd -`

3. 快速回到家目录的方法: `cd`

4. Linux各个命令(例如`pwd`、`ls`)到底装在哪?
- `which`命令可寻找Linux命令所在的路径
- 练习: `which pwd`、`which ls`、`which python`、`which jupyter`...

5. 为什么Linux常用命令可以不加绝对路径或者相对路径执行？
- Linux具备默认搜索路径，默认搜索路径记录在环境变量`PATH`中
- 练习: `echo $PATH`观察环境变量PATH的内容，'$'是Linux下的取值运算符
- 练习: `man echo`搜寻echo的帮助文件
- 如果大家有兴趣，未来可加入如何修改环境变量`PATH`(等)内容

- **Linux下执行Python脚本的第二种方法 -- 自动调用Python命令执行Python脚本(.py文件)**
1. 寻找`python`命令所在的路径: `which python`
    - 所寻路径为`/opt/tljh/user/bin/python`
2. 打开需要运行的Python脚本，例如helloworld.py
3. 在首行添加python路径: 格式为“`#!<路径>`”
4. 为文件添加**执行权限**: `chmod u+x helloworld.py` 
    - 注: 感兴趣的同学自行检索更改权限命令`man chmod`，或者利用百度微软等搜索引擎
5. 执行脚本: `./helloworld.py`
