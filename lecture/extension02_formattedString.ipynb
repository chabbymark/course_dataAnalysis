### 字符串进阶

- **如何连接两个字符串**
    - 如果我们要在Python里打开一个文件，比如说`/home/<user>/homework/practice1.ipynb`, 我们可能需要把这个路径保存到一个变量里
        - `path_file = /home/<user>/homework/practice1.ipynb`
    - 如果一个文件的完整路径是`/run/media/bzc/savio/scratch/Run/test_ibm/mul_src/wrfout_d07_2017-07-14_23:17:40.nc`，**你还会把路径写成一行吗？**
    - 一般而言，不同的目录代表不同的含义, 比如:
        - `/run/media/bzc/`代表用户
        - `savio/scratch/`代表服务器
        - `test_ibm/mul_src/`代表不同的模拟
        - `wrfout_d07_2017-07-14_23:17:40.nc`代表数据文件
    - 以后大家访问数据，会不会遇到需要更改不同子目录，不同数据文件的情况？--会

*目的: 当有需要更改路径中某一段时，**我们不更改其他段的内容**。*

连接字符串最简单的方法: `+`

```python
base_user = '/run/media/bzc/'
base_sever = 'savio/scatch'
base_sim = 'test_ibm/mul_src/'
base_file = 'wrfout_d07_2017-07-14_b23:17:40.nc'

path_file = base_user + base_sever + base_sim + base_file
print(path_file)
```

- **如何快速创造分割线**

```python
for ind in range(5):
    print('---------------------------------------------------------------------------')
    print('The current number is', ind)
print('---------------------------------------------------------------------------')
```

使用`*`快速复制一个字符串

```python
for ind in range(5):
    print('-'*80)
    print('The current number is', ind)
print('-'*80)
```

<font size=10> **注:看上去完全相同的运算符对于不同类型的数据意义可能完全不同，比如**
- 字符串的`+`表示连接不同字符串，而数值类型的`+`表示数学意义上的加法
- 字符串的`*`表示字符串的复制，而数值类型的`*`表示数学意义的乘法

- **如何使用format函数(格式化字符串)？**

```python
person1 = {"name": "张三", "gender": "男", "age":"20", "ID":"370705"}
person2 = {"name": "李四", "gender": "女", "age":"21", "ID":"370706"}
person3 = {"name": "王五", "gender": "男", "age":"20", "ID":"370707"}

print('name:', person1['name'], '; gender:', person1['gender'], '; age:', person1['age'], '; ID:', person1['ID'])
print('name:', person2['name'], '; gender:', person2['gender'], '; age:', person2['age'], '; ID:', person2['ID'])
print('name:', person3['name'], '; gender:', person3['gender'], '; age:', person3['age'], '; ID:', person3['ID'])
```

```python
fmt_person = "name: {:s}; gender: {:s}; age: {:s}; ID: {:s}"

print(fmt_person.format(person1['name'], person1['gender'], person1['age'], person1['ID']))
print(fmt_person.format(person2['name'], person2['gender'], person2['age'], person2['ID']))
print(fmt_person.format(person3['name'], person3['gender'], person3['age'], person3['ID']))
```

看上去好像没短多少，但是
- 不需要将不同类型的代码混合输入，比如变量名和字符串
- 如果需要改输出格式，只需要更改format的字符串

什么是format格式化字符串？(`fmt_person = "name: {:s}; gender: {:s}; age: {:s}; ID: {:s}"`)
- 在字符串中用`{:}`定义你想替换的字段，比如上一个例子中，我们有四个字段需要替换，分辨是姓名、性别、年龄、证件号。
    - `{:}`冒号左边是变量名或变量位置，可省略；冒号右边是输入变量的变量类型以及输出格式。
    - `{:s}`代表字符串类型(str)，省略变量名或位置
    - `{name:s}`代表字符串类型(str)，format函数输入变量名name，例如`fmt_person.format(name=person1['name'])`
    - `{0:s}`代表字符串类型(str)，format函数输入变量位置为第0位，例如`fmt_person.format(person1['name'], person1['gender'])`
    - 变量类型还有: `d`-整数, `f`-浮点数, `e`-指数等

*练习: 用格式化字符串重塑路径`/run/media/bzc/savio/scratch/Run/test_ibm/mul_src/wrfout_d07_2017-07-14_23:17:40.nc`*

```python
base_user = '/run/media/bzc/'
base_sever = 'savio/scatch/'
base_sim = 'test_ibm/mul_src/'
base_file = 'wrfout_d07_2017-07-14_b23:17:40.nc'

fmt_file = '{:s}'*4
print(fmt_file.format(base_user, base_sever, base_sim, base_file))
```
