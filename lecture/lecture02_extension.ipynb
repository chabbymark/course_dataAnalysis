### 数据类型

#### 基本数据类型

- 整型
    - 带符号(`num=1`/`num=-10`)
    - 不限长度(`num=2**1000`)

- 浮点型
    - 双精度
    - 截断误差

```python
print("1+1e-16等于1是否正确", 1+1e-16==1)
```

- 复数
    - 采用‘实部+虚部j’的方式定义，如`100+3j`

- **字符串**
    - 两种定义方式: 双引号`""`或单引号`''`
    - 支持多行字符串
    - 支持转义字符，如`\n`, `\t`
    - **如何格式化输出？**
        - `print('name:', person1['name'], '; gender:', person1['gender'], '; age:', person1['age'], '; ID:', person1['ID'])`<br>
        `print('name:', person2['name'], '; gender:', person2['gender'], '; age:', person2['age'], '; ID:', person2['ID'])`<br>
        ...

- 逻辑类型
    - `True`或者`False`

- 空值类型
    - `None`
