## 📖知识拓展--Linux文件权限(Permission)

### ❓为什么有些同学能看到我的作业，但我却看不了有些同学的作业

### 💡罪魁祸首: Linux文件具有不同权限(permission)。

- ### `ls -al`显示详细信息后可以观察到文件的权限。
- ### 用户群组分为三个级别: 用户(user)、组员(group)、其他(other)。
- ### 每个群组对应三个权限: 读(read)、写(write)、执行(excute)。
- ### 如果看到对应字母就代表有权限。

## 

### ❓如何更改权限

### 💡`chmod`命令

### 例如:
- ### `chmod g+x example1_practice.ipynb` -- 给example1_practice.ipynb的组员增加执行权限。
- ### `chmod go+rwx example1_practice.ipynb` -- 同时给组员和其他人增加读、写、执行权限。
- ### `chmod 750 example1_practice.ipynb` -- 还可以用数字表示三个用户群组的权限。
    - ### rwx对应二进制数111，对应十进制数7
    - ### 同理r-x对应二进制数101，对应十进制数5
    - ### 750即表示用户权限7(rwx)，组员权限5(r-x)，其他权限0(---)

# 🔍注: 自己的作业请改成'rwx------'（700）权限后再提交!!!
- ### `chmod go-rwx <文件名>`
- ### `chmod 700 <文件名>`
- ### 还有，提交文件的时候请用`cp`命令，应该不会出现权限问题。

## 

### ❓怎么改变一个目录下默认创建文件的权限 

### 💡`umask`命令 -- 创建文件时，去除何种权限。
- ### `umask` -- 不加任何参数时显示创建文件时去除何种权限。
- ### `umask 022` -- 修改创建文件时的默认权限，其中用户不去除任何权限，组会去除写权限、其他人去除写权限。

### 🔍注: 嫌每次用chmod改权限麻烦的同学可以修改自己homework目录的默认创建文件权限。
